```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("Student_Performance.csv")
df.head()
```

```python
sns.scatterplot(x="study_hours",y="overall_score",data=df)
plt.show()
```

```python
df=df.drop_duplicates()
df.columns=df.columns.str.strip()
df["age"]=df["age"].astype(int)
df.describe()



```

```python
df["gender"].unique()
```

df["gender"].value_counts().plot(kind="bar")
plt.title(Gender Distribution)
plt.show()

```python
df["gender"].value_counts().plot(kind="bar")
plt.title("gender distribution")
plt.xlabel("Gender")
plt.ylabel("Number of students")
```

```python
plt.hist(df["math_score"],bins=20)
plt.title("Math Score Distribution")
plt.xlabel("Math Score")
plt.ylabel("Number of Students")
plt.show()
```

```python
df[["math_score","science_score","english_score"]].mean().plot(kind="bar")
plt.title("Average scores subject wise")
plt.show()
```

```python
sns.boxplot(x="gender",y="math_score",data=df)
plt.title("Math score by gender")
plt.show()
```

```python
sns.heatmap(df[["math_score","science_score","english_score"]].corr(),annot=True)
plt.title("Correlation Between Subjects")
plt.show()
```

```python
df[["math_score","science_score","english_score"]].plot(kind="box")

plt.title("Score Distribution")

plt.show()
```

```python
sns.pairplot(df[["math_score","science_score","english_score"]])
plt.show()
```
